```python
from tkinter import *
from datetime import datetime
import os
from pathlib import Path
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import threading
import pytesseract
import pyttsx3
import cv2


###pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
engine = pyttsx3.init()

###__file__ = "../Datasets/"
###IMAGE_PATH = Path(__file__).parent/"res"

### Path to Store Captured Images
IMAGE_PATH = Path('/home/mansura/CSE_3200').parent / "res"

### Create res directory regardless of its existence
os.makedirs(IMAGE_PATH.resolve(), exist_ok=True)
is_speaking = False


def opencv_capture():
    cam = cv2.VideoCapture(0)
    cv2.namedWindow("Image Capture")
    img_counter = 0

    while True:
        ret, frame = cam.read()
        if not ret:
            print("failed to grab frame")
            break
        cv2.imshow("Image Capture", frame)

        k = cv2.waitKey(1)
        if k % 256 == 27:
            # ESC pressed
            print("Escape hit, closing...")
            break
        elif k % 256 == 32:
            # SPACE pressed
            img_name = "opencv_frame_{}.png".format(
                datetime.timestamp(datetime.utcnow())
            )
            filename = (IMAGE_PATH / img_name).resolve()
            filename = str(filename)
            cv2.imwrite(filename, frame)
            print("{} written!".format(filename))
            img_counter += 1
            upload_file(filename)

    cam.release()
    cv2.destroyAllWindows()
    # Exit from the thread
    exit(0)


def capture_image():
    ### Run on a separate thread to keep main thread (GUI) run without blocking
    th = threading.Thread(target=opencv_capture)
    th.start()


def speakInThread(txt):
    words = txt.split()
    for w in words:
        if is_speaking is False:
            break
        engine.say(w)
        engine.runAndWait()
    engine.stop()


def speak(txt):
    global is_speaking
    is_speaking = speakInThread
    th = threading.Thread(target=speakInThread, args=(txt,))
    th.start()


def stopSpeaking():
    global is_speaking
    is_speaking = False

### Upload File Function called by Button Press
def upload_file(filename=None):
    global img, img1

    if filename is None:
        fileType = [("Jpg File", "*.jpg"), ("PNG File", "*.png")]
        filename = filedialog.askopenfilename(filetypes=fileType)
    
    ### Upload Button Background Assignment in Function
    button1.image= image1
    img = Image.open(filename)
    ### Selected Image Size
    imgResized = img.resize((280, 200))
    img1 = ImageTk.PhotoImage(imgResized)
    ### Selected Image Placed in Button2
    button2 = tk.Button(window, image=img1, bg="#15666E")  ####
    button2.grid(row=3, column=1)
    ### Selected image button position
    button2.place(x=100, y=280)    
    txt = pytesseract.image_to_string(img) 
    ### Button3 to show text
    button3 = tk.Button(window, text=txt)
    button3.grid(row=3, column=1)
    ### Button3 Position
    button3.place(x=650, y=280)
    img3 = PhotoImage(file = f"img3.png")
    ### Button4 for Speak Button
    button4 = tk.Button(
        window,
        image= img3,
        text="SPEAK",
        command=lambda: speak(txt),
        font=interfaceFont1,
        activebackground='#222B36',
        borderwidth = 0,
        highlightthickness = 0,
        bg="#85929E",
        relief='flat'
    )
    button1.image = img3
    button4.grid(row=1, column=1)
    ### Button4 position
    button4.place(x=400, y=250)
    img2 = PhotoImage(file = f"img2.png")
    ### Same Button4 used for Stop in other postion
    button4 = tk.Button(
        window,
        image = img2,
        text="STOP",
        command=lambda: stopSpeaking(),
        font=interfaceFont1,
        activebackground='#222B36',
        borderwidth = 0,
        highlightthickness = 0,
        bg="#85929E",
        relief='flat'
    )
    button4.image = img2
    button4.grid(row=1, column=1)
    ### Stop Button Position
    button4.place(x=400, y=400)
    

window = Tk()

### Window Size of the Software
window.geometry("1200x980")
window.title("Image to Speech Converter")
interfaceFont = ("Arial", 20, "bold")
interfaceFont1 = ("Arial", 10, "bold")
window.configure(bg = "#1b1e27")
canvas = Canvas(
    window,
    bg = "#1b1e27",
    height = 650,
    width = 1000,
    bd = 0,
    highlightthickness = 0,
    relief = "ridge")
canvas.place(x = 0, y = 0)

### Background Image Source
background_img = PhotoImage(file = f"/home/mansura/CSE_3200/Img_Speech/background.png")
background = canvas.create_image(540.5, 324.5,image=background_img)
global image1   
image1 = PhotoImage(file = f"/home/mansura/CSE_3200/Img_Speech/img1.png")

### Upload File Button
button1 = tk.Button(
    window,    
    image = image1,
    borderwidth = 0,   
    activebackground='#222B36',
    text="Upload File",    
    highlightthickness= 0,     
    command=lambda: upload_file(),
    font=interfaceFont1,  
    bg="#15929E",
    relief='flat',
)
button1.image= image1
button1.grid(row=13, column=2)
### Upload File Button Position
button1.place(x=170, y=130)

### Capture IMage Button
img0 = PhotoImage(file = f"/home/mansura/CSE_3200/Img_Speech/img0.png")
captureBtn = tk.Button(
    window,
    image = img0,
    borderwidth = 0,
    activebackground='#222B36',
    text="Capture",    
    highlightthickness = 0,
    command=lambda: capture_image(),
    font=interfaceFont1,
    bg="#85929E",
    relief='flat'
)
captureBtn.image = img0
captureBtn.grid(row=3, column=2)
### Capture Button Position
captureBtn.place(x=630, y=130)

window.resizable(False, False)
window.mainloop()
```
